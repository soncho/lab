```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
def exp(n):
    return (1 + 1 / n) ** n
```

```python
plt.plot([exp(n) for n in np.arange(1, 100)])
plt.show()
```

```python
# ToDo: 数式直す
import math
def N(x, u, s):
    return (1 / math.sqrt(2 * math.pi * s)) * (math.e ** -((x - u) ** 2 / (2 * s )))
```

```python
plt.plot([N(x, 0, 1) for x in np.arange(-5, 5, 0.1)])
plt.show()
```
